## Проверка отчетов

### ClickHouse

```python
import clickhouse_connect

# === 1. Подключение к ClickHouse ===
client = clickhouse_connect.get_client(
    host="clickhouse", port=8123, username="custom_user", password="custom_password"
)

# === 2. Список витрин ===
report_tables = [
    "report_products",
    "report_customers",
    "report_time",
    "report_stores",
    "report_suppliers",
    "report_quality",
]

# === 3. Выгрузка и отображение ===
for table in report_tables:
    print(f"\n📄 {table} — preview:")
    df = client.query_df(f"SELECT * FROM {table} ORDER BY 1 LIMIT 20")
    print(df)
```
